### Best Picture Table from Wikipedia

```python
#import library to query webpage of interest
import urllib.request
#specifying page of interest
wiki = "https://en.wikipedia.org/wiki/Academy_Award_for_Best_Picture"
    
#save the HTML of the site within the page variable
page = urllib.request.urlopen(wiki)

#import library to parse HTML from page
from bs4 import BeautifulSoup

#parse data from "page" and save to new variable "soup"
soup = BeautifulSoup(page, 'lxml')
```

```python
#pinpointing the location of the table and its contents
tables = soup.findAll("table", class_ = "wikitable")

#creating lists for each of the columns I know to be in my table.
films=[]
studios=[]

#utilizing HTML tags for rows <tr> and elements <td> to 
# iterate through each row of data and append data elements to their appropriate lists:
for table in tables:
    for row in table.findAll("tr"):
        cells = row.findAll("td")
        if len(cells)==2: #Only extract table body not heading
            films.append(cells[0].find(text=True).strip())
            studios.append(cells[1].find(text=True).strip())
```
